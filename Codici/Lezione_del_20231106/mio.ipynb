```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# 
dati = np.loadtxt('Data_altezze.dat')
display(dati)
```

```python
#Calcolare la media e deviazione standard 
media = np.mean(dati)
deviazione_standard = np.std(dati)

display(dati, deviazione_standard)
```

```python
#Istogramma
plt.subplots(1,1)
plt.hist(dati, bins=20, edgecolor = "black")

plt.show()
```

Distribuzione Normale

```python
from scipy.stats import skew
from scipy.stats import kurtosis
```

```python
s = skew(dati)
k = kurtosis(dati, fisher=False)

display(s, k)
```

```python
#PlotBox
plt.subplots(1,1)
plt.boxplot(dati)

plt.show()
```

```python
#QQ PLOT
from scipy.stats import probplot
from scipy.stats import uniform
fig, ax = plt.subplots(1,1)
probplot(dati, dist = uniform, plot = ax)
plt.show()
```

```python
#Grafico a torta 
modalita = ["automobile", "a piedi", "autobus", "scooter", "bicicletta"]
frequenza = [7,3,9,4,5]
fig, ax = plt.subplots(1,1)

#ax.pie()  la funzione che crea il grafico a torta 

ax.pie(frequenza, labels = modalita, autopct ='%1.1f%%' )
plt.show()

```
